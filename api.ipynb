```python
temp_info = []
more_data = []
for i in range (len(data)):
    # print('Planet name: ', data[i].get('name'))
    # print('Player Count: ', data[i]['statistics'].get('playerCount'))
    # print('Current owner: ', data[i].get('currentOwner'))
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M")
    planet_name = data[i].get('name')
    player_count = data[i]['statistics'].get('playerCount')
    owner = data[i].get('currentOwner')
    temp_info.append( [current_time, planet_name, player_count, owner])
    # print(temp_info)
    is_disabled = data[i].get('disabled')
    mission_success_rate = data[i]['statistics'].get('missionSuccessRate')
    current_kda = 0
    try:
        current_kda = data[i]['statistics'].get('terminidKills') + data[i]['statistics'].get('automatonKills') / data[i]['statistics'].get('deaths')
    except: 
        current_kda = 0
    total_deaths = data[i]['statistics'].get('deaths')
    more_data.append( [current_time, planet_name, player_count, owner, is_disabled, mission_success_rate, current_kda, total_deaths])

    # print('-----------------------------')

more_data
```

```python
df = pd.DataFrame(temp_info)
df.to_csv('warData.csv', index = False, header = False, mode = 'a')
```
